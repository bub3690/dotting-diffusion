```python
import requests
import json

```

#   txt2img

```python
dat = json.dumps({"prompt": "men riding brown horse"})
res = requests.post("http://localhost:3000/txt2img/", data=dat)
```

```python
print(res.headers)

# print response keys
print(res)
```

```python

#res.content

# get image1, image2 from res
```

```python
from PIL import Image
import io   
image = Image.open(io.BytesIO(res.content))
```
