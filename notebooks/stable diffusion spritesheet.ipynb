```python
from diffusers import DiffusionPipeline

pipeline = DiffusionPipeline.from_pretrained("Onodofthenorth/SD_PixelArt_SpriteSheet_Generator")
```

```python
pipeline
```

```python
pipeline.to("cuda")
prompt = "man holding sword,PixelartFSS"
#PixelartFSS, PixelartRSS, PixelartBSS, PixelartLSS
image = pipeline(prompt,
                 guidance_scale=7.5,
                 negative_prompt=",".join(['']),
                 num_images_per_prompt=1,
                 num_inference_steps=50,
                 ).images[0]


#image=[remove(sample) for sample in image ]
```
